```python
#--- Import Pandas ---
import pandas as pd

#--- Read in dataset ----
df = pd.read_csv("lung_cancer_examples.csv")

df
```

```python

duplicates = df.duplicated().sum()

duplicates
```

```python
null_values = df.isnull().sum()

null_values
```

```python
df = df.rename(columns={"Alkhol":"Alcohol"})
df
```

```python
value_smokes = df["Smokes"].value_counts()

value_smokes
```

```python
def categorize_smokers(x):
    if x == 0:
        return 'Non-Smokers'
    elif x <= 2:
        return 'Light Smokers'
    elif x > 2 and x <= 10:
        return 'Mediocre Smokers'
    else:
        return 'Heavy Smokers'


df["Smoking_Category"] = df["Smokes"].apply(categorize_smokers)

df
```

```python
value_alcohol = df["Alcohol"].value_counts()

value_alcohol
```

```python
def categorize_alcohol(x):
    if x == 0:
        return 'Non-Drinkers'
    elif x <= 2:
        return 'Light Drinkers'
    elif x > 2 and x <= 10:
        return 'Mediocre Drinkers'
    else:
        return 'Heavy Drinkers'
df["Alcohol_Category"] = df["Alcohol"].apply(categorize_alcohol)
df
```

```python
df.to_csv("lung_cancer.csv", index= False)
df
```
